#Demanda Cobb Douglas

\begin{equation}
x_1 = \frac{\alpha_1}{\alpha_1 + \alpha_2} \frac{m}{p_1}
\end{equation}

```python
# Este es un programa de python que grafica curvas de demanda

# Cargamos bibliotecas de matrices y de gráficos
# Seaborn no es estrictamente necesario. Se usa para la estética de los gráficos
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

# Definimos la función de demanda Cobb Douglas
# La llamamos "f1"
# En esta función hay parámetros: alpha1 y alpha2
# y hay argumentos: p1, p2, m
def f1(p1,p2,m):
  return alpha1/(alpha1+alpha2)*m/p1

# Damos valores a los parámetros
alpha1 = 0.4
alpha2 = 0.6

# A modo de ejemplo evaluamos la función de demanda en p1=2 p2=2 m=100
q1=f1(2,2,100)
print(q1)
```

```python
# Creamos un vector de precios p1
p1=np.array([1, 2, 3, 4, 5])
print(p1)
```

```python
# Evaluamos la función de demanda para cada elemento del vector de precios p1
q1=f1(p1,2,100)
print(q1)
```

```python
from matplotlib.markers import MarkerStyle
# Graficamos la demanda a partir de los vectores q1 y p1
plt.plot(q1, p1)
plt.plot(q1, p1, 'o')
plt.show
```

```python
# Ahora vamos a crear vectores más grandes para que el gráfico se vea mejor
# Valor mínimo: 0.5; Valor máximo: 6; Número de elementos: 150
p1=np.linspace(0.5,6,150)
print(p1)
```

```python
# Evaluamos la demanda para cada punto
print("Evaluando...")
q1=f1(p1,2,100)
print("Listo")
```

```python
#Con el nuevo vector de precios, con mayor numero de elementos, graficamos la demanda.
plt.figure(figsize=(10,5), dpi=100)
plt.plot(f1(p1,2,100),p1, linewidth=3, color='darkgreen')
#Agregamos nombres a los ejes y al grafico
plt.xlabel("Q1")
plt.ylabel("P1")
plt.title('Demanda del bien 1', fontsize=15, fontstyle='oblique')
#Usamos el paquete seaborn para cuestiones estéticas
sns.despine()
#Mostramos el gráfico en pantalla
plt.show
# Guardamos el gráfico en un archivo
#la opcion dpi=300 mejora la resolucion del grafico cuando se descarga como png
plt.savefig('demanda.png', dpi=300)
```

#Demanda Coeficientes fijos
\begin{equation}
x_1^*=\frac{um}{up_1+vp_2}
\end{equation}

```python
# Funcion que crea la demanda por el bien 1
def f2(p1,p2,m):
  return (u*m)/(u*p1+v*p2)

# Damos valores a los parámetros
u = 2
v = 3
```

```python
# Para p1 armamos un vector de precios para evaluar la demanda
p1=np.linspace(0,6,150)

# Dado p2=2 y m=100 nos guardamos en la variable q la demanda del bien 1 para cada nivel de precios p1
q1=f2(p1,2,100)

# Graficamos la demanda
plt.figure(figsize=(8,5))
plt.plot(q1,p1, linewidth=3, color='orange')
plt.xlabel('X1')
plt.ylabel('P1')
plt.title('Demanda con preferencias de coeficientes fijos', fontsize=16)
sns.despine()
plt.show
plt.savefig('coeficientesfijos.png',dpi=300)
```

#Demanda con preferencias lineales
$x_1^*(p1,p2,m)=
\begin{cases}
m/p_1 & si~a_1/a_2 \geq p_1/p_2 \\
0 & si~a_1/a_2 > p_1/p_2
\end{cases}
$

```python
# Demanda del bien 1
def f3(p1,p2,m):
  q=np.where(a1/a2>=p1/p2,m/p1,0)
  return q

# Seteamos los parametros
a1 = 2
a2 = 3

# Generamos vector de precios
p1=np.linspace(0.5,7,150)

# Evaluamos la demanda en el vector de precios dado p2=4 y m=100
q=f3(p1,4,100)
```

```python
# font del titulo
csfont = {'fontname':'Times New Roman'}

# Grafico
plt.figure(figsize=(8,5), dpi=100)
plt.plot(q,p1, linewidth=3, color='blue')
plt.axis([-1,100,0,7])
plt.title('Demanda con preferencias lineales', fontsize=16)
sns.despine()
plt.show
```

#¿Qué hace np.where?


Esto lo que hace es evaluar elemento por elemento del vector si se cumple la condicion. 
Si $a_1/a_2\geq p_1/p_2$ para un $p_1$ del vector entonces pone $m/p_1$ en el vector de cantidades.
Si $a_1/a_2< p_1/p_2$ entonces pone un 0

Tal como lo dice la demanda

#Curvas de indiferencia

Dada una funcion de utilidad Cobb-Douglas de la forma $u(x_1,x_2)=x_1^{\alpha_1}x_2^{\alpha_2}$


```python
# Funcion de utilidad Cobb-Douglas
# El doble asterisco significa exponente
def g(x1,x2):
  return x1**alpha1*x2**alpha2

# Definimos los parametros
alpha1 = 0.5
alpha2 = 0.5

# Tenemos que darle valores a x1 y x2
#Lo podemos hacer a mano como una lista o como vimos usando np.linspace
x1=np.linspace(0,10,200)
x2=x1

#Con el comando np.meshgrid generamos el par de valores correspondientes a x1,x2, necesario para cuando hacemos graficos en 3d y curvas de nivel
x1,x2=np.meshgrid(x1,x2)
```

```python
# Evaluamos la funcion en todos los valores de x1 y x2
z=g(x1,x2) 

# Elegimos el estilo de grafico
# Ver todos en http://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html

# Ejecutamos el gráfico
plt.figure(figsize=(8,5),dpi=100)
CI = plt.contour(x1,x2,z, levels=[1,1.3,1.6], colors=['gray','cyan','green'])
#El comando contour permite graficar las curvas de nivel.
#El parametro levels es opcional, y corresponde a los niveles de z a lo largo de las curvas
plt.xlabel('x1')
plt.ylabel('x2')
plt.title("Cobb-Douglas alpha=0.5, beta=0.5 ")
#Definimos los limites de los ejes, el eje de las abcisas de 0 a 5 y de las ordenadas de 0 a 5
plt.axis([0,5,0,5])
#Definimos la estética del gráfico
sns.despine()
plt.grid(True)
```

Sea la funcion de utilidad 

\begin{equation}
u(x_1,x_2)=min\{\frac{x_1}{a},\frac{x_2}{b}\}
\end{equation}

```python
#Haciendo lo mismo que antes pero modificando la funcion de utilidad tenemos
def h(x1,x2):
  return np.fmin(x1/a,x2/b) #la funcion np.fmin devuelve el minimo valor.
```

```python
# Definimos parametros
a = 2
b = 3

# Vector de demandas
x1=np.linspace(0,50,100)
x2=x1
x1,x2=np.meshgrid(x1,x2)

# Curva de indiferencia
z=h(x1,x2)

# Grafico
plt.figure(figsize=(8,5), dpi=100)
plt.axis([0,55,0,55])
plt.contour(x1,x2,z, levels=[3,6,12], linewidths=3)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title("Coeficientes fijos", fontsize=16)
sns.despine()
plt.show
```

Para una funcion de preferencias lineales
\begin{equation}
u(x_1,x_2)=a_1x_1+a_2x_2
\end{equation}


```python
# Funcion de utilidad de pref lineales
def h1(x1,x2):
  return a1*x1+a2*x2

# Definimos los parametros
a1 = 1
a2 = 2

# Vectores de demandas
x1=np.linspace(0,50,100)
x2=x1
x1,x2=np.meshgrid(x1,x2)

# Curva de indiferencia
z=h1(x1,x2)

# Grafico
plt.figure(figsize=(8,5), dpi=100)
plt.contour(x1,x2,z,levels=[30,40,50], linewidths=3)
plt.title("Preferencias lineales", fontsize=16)
sns.despine()
plt.show
```
